## Bài tập 7: Sử dụng PCA để biểu diễn các bài hát bằng biểu đồ scatter (dùng dữ liệu featuresdf). So sánh với việc sử dụng TSNE để giảm chiều (giải thích các tham số trong TSNE)

```python
# Loại bỏ các cột không cần thiết
X = feature_df.drop(['name', 'artists'], axis=1)
# Chuẩn hóa các thuộc tính
X = (X - X.mean()) / X.std()

from sklearn.impute import SimpleImputer

# Xử lý giá trị NaN 
# Thay thế NaN bằng giá trị trung bình của cột đó
imp = SimpleImputer(strategy='mean')
X = imp.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

# Sử dụng PCA để giảm chiều dữ liệu
pca = PCA(n_components=2)
pca.fit(X)
X_pca = pca.transform(X)

# Tạo biểu đồ scatter
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

```python
from sklearn.manifold import TSNE

# Sử dụng TSNE để giảm chiều dữ liệu
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200)
X_tsne = tsne.fit_transform(X)

# Tạo biểu đồ scatter
plt.scatter(X_tsne[:, 0], X_tsne[:, 1])
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

### Những tham số trong thuật toán TSNE bao gồm:

- n_components: Số chiều của không gian mới. Trong bài, ta sử dụng n_components=2 để có thể biểu diễn dữ liệu bằng các biểu đồ scatter 2 chiều.

- perplexity: Xác định kích thước của khu vực xung quanh mỗi điểm trong không gian gốc được xem xét để tính toán xác suất điểm lân cận. Thường thì perplexity thuộc [5, 50]. Ở đây, ta dùng perplexity = 30

-  learning_rate: Tốc độ học của thuật toán. Nếu learning_rate quá lớn, dữ liệu có thể tràn sang những khu vực không mong muốn của không gian mới. Ngược lại, việc hội tụ có thể rất chậm. learning_rate thường được đặt [10, 1000]. Ta dùng learning_rate = 200.

### So sánh với việc sử dụng TSNE để giảm chiều

PCA và TSNE đều là các phương pháp giảm chiều dữ liệu để dễ trực quan hóa và phân tích. 
 
Trong đó:

+ PCA giữ lại những thông tin quan trọng nhất về sự phân tán của dữ liệu, còn TSNE tập trung vào sự tương đồng giữa các điểm dữ liệu gốc trong không gian mới.

+ Các tham số trong PCA tương đối đơn giản và dễ hiểu, TSNE thì phức tạp hơn và ta cần đặt giá trị phù hợp để kết quả hiển thị hợp lí.

+ Điểm mạnh của TSNE là thường tạo ra các biểu đồ scatter tốt hơn so với PCA, nhất là khi dữ liệu có cấu trúc phức tạp và non-linear. Tuy nhiên, TSNE yêu cầu nhiều tài nguyên tính toán hơn so với PCA và thời gian tính toán có thể lâu hơn.

Kết luận: Nếu dữ liệu có cấu trúc đơn giản và tuyến tính, PCA có thể là lựa chọn tốt. Ngược lại, nếu dữ liệu có cấu trúc phức tạp và non-linear, hoặc khi bạn cần tạo ra các biểu đồ scatter đẹp hơn, TSNE có thể là lựa chọn phù hợp hơn.
