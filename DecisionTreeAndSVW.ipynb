```python
import pandas as pd
# Đọc dữ liệu từ tập tin CSV
df = pd.read_csv('https://drive.google.com/uc?id=11Ibd34CRcY9suKnpAFlWvy8o4-Dr5IPL')
```

### **Thay toàn bộ giá trị null bằng giá trị trung bình của cột**


---

Cú pháp: s.fillna(s.mean())


```python
# Xử lý cột Temperature
temp_max = df['Temperature'].value_counts().idxmax()  # Tần số xuất hiện nhiều nhất
df['Temperature'].fillna(temp_max, inplace=True)     # Điền giá trị thiếu bằng tần số max

# Điền giá trị trung vị cho các giá trị thiếu trong cột "Wind speed"
# Xử lý cột Wind Speed
wind_median = df['Wind speed'].median()               # Trung vị
df['Wind speed'].fillna(wind_median, inplace=True)    # Điền giá trị thiếu bằng trung vị
df
```

```python
transformed_data = pd.get_dummies(df[["Outlook","Temperature","Humidity","Wind speed"]])
transformed_data
```

```python
x_data = transformed_data.values
y_data = df["Play cricket"].values
```

```python
from sklearn.model_selection import train_test_split
x_data_train, x_data_test, y_data_train, y_data_test = train_test_split(x_data, y_data,test_size=0.2,random_state=1 )
```

## Decision Tree
- **max_depth**: là một tham số kiểm soát độ sâu của cây. Nó xác định số lượng tối đa các nút con được phép trong một nút trên cây. Nếu max_depth quá lớn, mô hình sẽ quá phức tạp và có khả năng overfitting cao, ngược lại, nếu max_depth quá nhỏ, mô hình sẽ quá đơn giản và có khả năng underfitting cao.
- **min_samples_leaf**: là một tham số kiểm soát số lượng mẫu tối thiểu được yêu cầu để một nút có thể trở thành lá. Tham số này giúp kiểm soát độ sâu của cây và tránh tình trạng overfitting. Nếu min_samples_leaf quá nhỏ, mô hình sẽ quá phức tạp và có khả năng overfitting cao, ngược lại, nếu min_samples_leaf quá lớn, mô hình sẽ quá đơn giản và có khả năng underfitting cao.

## SVM
- **kernel**: là một hàm toán học dùng để chuyển đổi dữ liệu từ không gian gốc sang không gian mới để dữ liệu có thể phân chia tuyến tính hoặc phi tuyến. Có nhiều loại kernel khác nhau, trong đó kernel tuyến tính (linear), polynomial (poly), Gaussian RBF (rbf), và sigmoid là những kernel phổ biến nhất.
- **C**: là một tham số điều khiển sự phân chia của mô hình. Tham số này xác định độ lớn của các margin và giúp kiểm soát việc overfitting hoặc underfitting. Càng lớn thì mô hình càng phức tạp và có khả năng overfitting cao, ngược lại thì mô hình càng đơn giản và có khả năng underfitting cao.

```python

from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# SVM
# Tìm các giá trị tốt nhất cho các tham số kernel và C
best_svm = None
best_svm_score = 0
for kernel in ['linear', 'poly', 'rbf', 'sigmoid']:
    for C in [0.1, 1, 10, 100]:
        svm = SVC(kernel=kernel, C=C)
        svm.fit(x_data_train, y_data_train)
        score = accuracy_score(y_data_test, svm.predict(x_data_test))
        if score > best_svm_score:
            best_svm = svm
            best_svm_score = score
print("Best SVM: ", best_svm)
print("SVM accuracy score: ", best_svm_score)

# Decision Tree
# Tìm các giá trị tốt nhất cho các tham số max_depth và min_samples_leaf
best_dt = None
best_dt_score = 0
for max_depth in [2, 3, 4, 5]:
    for min_samples_leaf in [1, 2, 3, 4]:
        dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        dt.fit(x_data_train, y_data_train)
        score = accuracy_score(y_data_test, dt.predict(x_data_test))
        if score > best_dt_score:
            best_dt = dt
            best_dt_score = score
print("Best Decision Tree: ", best_dt)
print("Decision Tree accuracy score: ", best_dt_score)
```
